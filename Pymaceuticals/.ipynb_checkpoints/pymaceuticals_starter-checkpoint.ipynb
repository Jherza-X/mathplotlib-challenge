```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from sklearn.linear_model import LinearRegression

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_data_complete = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
mouse_data_complete.head()
```

```python
# Checking the number of mice.

mouse_count_unique = mouse_data_complete["Mouse ID"].nunique()
mouse_count_unique
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = mouse_data_complete[mouse_data_complete.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]

unique_duplicate_mice_ids = duplicate_mice['Mouse ID'].unique()
unique_duplicate_mice_ids
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

#Mean calculation and column name change in a data frame
tumor_vol_mean = mouse_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean().reset_index()
tumor_vol_mean.rename(columns={"Tumor Volume (mm3)":"Tumor Vol Mean"}, inplace=True)

#tumor_vol_mean
#Median calculation and column name change in a data frame
tumor_vol_median = mouse_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].median().reset_index()
tumor_vol_median.rename(columns={"Tumor Volume (mm3)":"Tumor Vol Median"}, inplace=True)

#tumor_vol_median

#Variance calculation and column name change in a data frame
tumor_vol_variance = mouse_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].var().reset_index()
tumor_vol_variance.rename(columns={"Tumor Volume (mm3)":"Tumor Vol Variance"}, inplace=True)

#tumor_vol_variance
#Standar Deviation calculation and column name change in a data frame
tumor_vol_std_dev = mouse_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].std().reset_index()
tumor_vol_std_dev.rename(columns={"Tumor Volume (mm3)":"Tumor Vol STD"}, inplace=True)

#tumor_vol_std_dev
#SEM calculation and column name change in a data frame
tumor_vol_sem = mouse_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].apply(st.sem).reset_index()
tumor_vol_sem.rename(columns={"Tumor Volume (mm3)":"Tumor Vol SEM"}, inplace=True)
#tumor_vol_sem

# Merge all the statistics into one DataFrame
per_regimen_summary_stats = pd.merge(tumor_vol_mean, tumor_vol_median, on='Drug Regimen')
per_regimen_summary_stats = pd.merge(per_regimen_summary_stats, tumor_vol_variance, on='Drug Regimen')
per_regimen_summary_stats = pd.merge(per_regimen_summary_stats, tumor_vol_std_dev, on='Drug Regimen')
per_regimen_summary_stats = pd.merge(per_regimen_summary_stats, tumor_vol_sem, on='Drug Regimen')

per_regimen_summary_stats = per_regimen_summary_stats.set_index("Drug Regimen")
per_regimen_summary_stats

```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

#gruoupby - > group by Drug Regimen and then take Tumor Volume as value. 
#agg - > is aggregation fuction to apply one or more fuctions to a data frame column, we name each column in the new data frame   
each_regimen_stats = mouse_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(
    Mean=np.mean, Median=np.median, Variance=np.var, StdDev=np.std,
   SEM=st.sem 
).reset_index()

each_regimen_stats = each_regimen_stats.set_index("Drug Regimen")
each_regimen_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoints = mouse_data_complete.groupby('Mouse ID')['Timepoint'].max().reset_index()
#max_timepoints


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume = pd.merge(max_timepoints, mouse_data_complete, on=['Mouse ID', 'Timepoint'], how='left')
final_tumor_volume
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
# Filter for the desired treatment regimens: Capomulin, Ramicane, Infubinol, and Ceftamin
final_tumor_volume = final_tumor_volume[final_tumor_volume['Drug Regimen'].isin(treatment_regimens)]
final_tumor_volume
```

```python

# Create empty list to fill with tumor vol data (for plotting)
vol_data_plt = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
for regimen in treatment_regimens:
    regimen_data = final_tumor_volume[final_tumor_volume['Drug Regimen'] == regimen]
    Q1 = regimen_data['Tumor Volume (mm3)'].quantile(0.25)
    Q3 = regimen_data['Tumor Volume (mm3)'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    potential_outliers = regimen_data[(regimen_data['Tumor Volume (mm3)'] < lower_bound) | (regimen_data['Tumor Volume (mm3)'] > upper_bound)]
    # add subset 

    
    # Determine outliers using upper and lower bounds
    
    print(f"Results for {regimen}:\nIQR: {IQR}\nLower Bound: {lower_bound}\nUpper Bound: {upper_bound}")
    print("Potential Outliers:")
    print(potential_outliers)
    
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
# Group the data by 'Drug Regimen'
grouped_data = final_tumor_volume.groupby('Drug Regimen')['Tumor Volume (mm3)'].apply(list).reset_index()


# Create a box plot
plt.figure(figsize=(6, 4))
plt.boxplot(grouped_data['Tumor Volume (mm3)'], labels=grouped_data['Drug Regimen'], vert=True)
plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Treatment Group')
plt.title('Distribution of Tumor Volume for Each Treatment Group')
plt.tight_layout()

# Display the box plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
#from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
# Calculate the correlation coefficient
#correlation_coefficient, _ = pearsonr(capomulin_data['Weight (g)'], capomulin_data['Tumor Volume (mm3)'])
correlation_coefficient, _ = pearsonr(merged_data['Weight (g)'], merged_data['Average Tumor Vol'])
print(f'Correlation Coefficient: {correlation_coefficient}')
```

```python
# Create a scatter plot for mouse weight vs. average tumor volume
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Weight (g)'], merged_data['Average Tumor Vol'], marker='o', label='Data Points')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.grid(True)

# Create a linear regression model
from scipy.stats import linregress
# Perform linear regression using linregress
slope, intercept, rvalue, pvalue, stderr = linregress(merged_data['Weight (g)'], merged_data['Average Tumor Vol'])

# Create the linear equation string
line_eq = f'y = {slope:.2f}x + {intercept:.2f}'

# Predict values using the linear regression model
y_pred = slope * merged_data['Weight (g)'] + intercept

# Plot the regression line
plt.plot(merged_data['Weight (g)'], y_pred, color='red', linewidth=2, label='Linear Regression')

# Annotate the plot with the linear equation
plt.annotate(line_eq, (22, 36), fontsize=15, color="red")

# Display the legend
plt.legend()

# Display the scatter plot with the regression line
plt.tight_layout()
plt.show()



#########Diiferent Way of doing
# X = merged_data['Weight (g)'].values.reshape(-1, 1)
# y = merged_data['Average Tumor Vol'].values

# #(slope, intercept, rvalue, pvalue, stderr) = linregress(X, y)
# #regress_values = X * slope + intercept
# #line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# reg = LinearRegression().fit(X, y)

# # Predict values using the linear regression model
# y_pred = reg.predict(X)

# # Plot the regression line
# plt.plot(X, y_pred, color='red', linewidth=2, label='Linear Regression')
# #plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
# # Display the legend
# plt.legend()

# # Display the scatter plot with the regression line
# plt.tight_layout()
# plt.show()
```
